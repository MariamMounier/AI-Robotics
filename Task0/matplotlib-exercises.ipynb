```python
# Import NumPy as np
import numpy as np
```

```python
# Create an array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X
X = np.linspace(0, 100, 100)
```

```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)

# Calculate X^2
X_squared = X ** 2

# Create a figure and a single subplot
fig, ax = plt.subplots()

# Plot X versus X^2
ax.plot(X, X_squared, label='X^2')

# Add labels and a title
ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('Plot of X versus X^2')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

We'll start with scatter plots.

```python
# Create a scatter plot of X versus the exponential of X (np.exp(X))

# Calculate the exponential of X
exp_X = np.exp(X)

# Create a scatter plot of X versus exp(X)
plt.scatter(X, exp_X, marker='o', color='blue', label='X vs. exp(X)')

# Add labels and a title
plt.xlabel('X')
plt.ylabel('exp(X)')
plt.title('Scatter Plot of X vs. exp(X)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Create a scatter plot of X versus np.sin(X)
X = np.linspace(0, 2 * np.pi, 100)

# Calculate the sine of X
sin_X = np.sin(X)

# Create a scatter plot of X versus sin(X)
plt.scatter(X, sin_X, marker='o', color='green', label='X vs. sin(X)')

# Add labels and a title
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.title('Scatter Plot of X vs. sin(X)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)


# Set the figsize for the entire figure
fig = plt.figure(figsize=(10, 5))

# Create subplots
ax1 = fig.add_subplot(221)  # Top left
ax2 = fig.add_subplot(222)  # Top right
ax3 = fig.add_subplot(223)  # Bottom left
ax4 = fig.add_subplot(224)  # Bottom right

# Adjust spacing between subplots to prevent overlap
plt.tight_layout()

# Plot X versus X/2 on the top left axes
X = np.linspace(0, 10, 100)
ax1.plot(X, X / 2)
ax1.set_title('X vs. X/2')

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
x_data = np.random.rand(10)
y_data = np.random.rand(10)
ax2.scatter(x_data, y_data)
ax2.set_title('Scatter Plot of 10 Random Numbers')

# Plot a bar graph of the favourite food keys and values on the bottom left subplot
ax3.bar(foods, prices)
ax3.set_title('Favorite Foods and Their Prices')
ax3.set_xlabel('Food Name')
ax3.set_ylabel('Price')

# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
X_norm = np.random.normal(0, 1, 1000)
ax4.hist(X_norm, bins=30, edgecolor='black')
ax4.set_title('Histogram of Normally Distributed Numbers')
ax4.set_xlabel('Value')
ax4.set_ylabel('Frequency')
# Show the figure
plt.show()
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
car_sales  =   pd.read_csv('C:/Users/Mariam/Downloads/robotics/Task0/data/car-sales.csv')
print(car_sales)
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales['Price'].plot()

# Show the plot
plt.show()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).str[:-2].astype(float)

# Display the updated DataFrame
print(car_sales)
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales['Sales Date'] = pd.date_range(start='2023-10-13', periods=len(car_sales))

# View the car_sales DataFrame
car_sales
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column

# Plot the 'Sale Date' column versus the 'Total Sales' column
plt.figure(figsize=(10, 5))
plt.plot(car_sales['Sales Date'], car_sales['Total Sales'])
plt.title('Total Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert the 'Price' column to the integers

car_sales['Price'] = car_sales['Price'].astype(int)

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function

# Create a scatter plot of 'Odometer (KM)' versus 'Price'
car_sales.plot(kind='scatter', x='Odometer (KM)', y='Price')

# Add labels and a title to the plot
plt.title('Scatter Plot of Odometer vs. Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')

# Show the plot
plt.show()
```

```python
# Create a NumPy array of random numbers of size (10, 4) and save it to X


X = np.random.rand(10, 4)

# Turn the NumPy array X into a DataFrame with columns ['a', 'b', 'c', 'd']
df = pd.DataFrame(X, columns=['a', 'b', 'c', 'd'])

# Create a bar graph of the DataFrame
df.plot(kind='bar')
plt.title('Bar Graph of DataFrame')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame

make_odometer = car_sales.groupby('Make')['Odometer (KM)'].mean()
make_odometer.plot(kind='bar')

# Add labels and a title to the plot
plt.title('Average Odometer Reading by Make')
plt.xlabel('Make')
plt.ylabel('Average Odometer (KM)')

# Show the plot
plt.show()
```

```python
# Create a histogram of the 'Odometer (KM)' column
car_sales['Odometer (KM)'].plot(kind='hist', bins=20, edgecolor='black')

# Add labels and a title to the plot
plt.title('Histogram of Odometer (KM)')
plt.xlabel('Odometer (KM)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Create a histogram of the 'Price' column with 20 bins
car_sales['Price'].plot(kind='hist', bins=20, edgecolor='black')

# Add labels and a title to the plot
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the following plot:

<img src="../images/matplotlib-heart-disease-chol-age-plot.png"/>

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old

patients_over_50 = heart_disease[heart_disease['age'] > 50]

# Calculate the mean age and cholesterol for all patients over 50
mean_age = patients_over_50['age'].mean()
mean_chol = patients_over_50['chol'].mean()

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data for patients with no heart disease (target=0)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 0]['age'],
    y=patients_over_50[patients_over_50['target'] == 0]['chol'],
    c='blue',
    label='No Heart Disease',
    alpha=0.7
)

# Plot the data for patients with heart disease (target=1)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 1]['age'],
    y=patients_over_50[patients_over_50['target'] == 1]['chol'],
    c='red',
    label='Heart Disease',
    alpha=0.7
)

# Customize the plot
plt.title('Age vs. Cholesterol for Patients Over 50')
plt.xlabel('Age')
plt.ylabel('Cholesterol (chol)')
plt.legend()

# Add mean lines
plt.axvline(mean_age, color='green', linestyle='--', label=f'Mean Age: {mean_age:.2f}')
plt.axhline(mean_chol, color='green', linestyle='--', label=f'Mean Chol: {mean_chol:.2f}')

# Remove grid lines
ax = plt.gca()
ax.grid(False)

# Show the plot
plt.legend()
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot
mean_age = patients_over_50['age'].mean()
mean_chol = patients_over_50['chol'].mean()

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data for patients with no heart disease (target=0)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 0]['age'],
    y=patients_over_50[patients_over_50['target'] == 0]['chol'],
    c='blue',
    label='No Heart Disease',
    alpha=0.7,
    cmap='winter'  # Change the cmap parameter to 'winter'
)

# Plot the data for patients with heart disease (target=1)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 1]['age'],
    y=patients_over_50[patients_over_50['target'] == 1]['chol'],
    label='Heart Disease',
    alpha=0.7,
    cmap='winter'  # Change the cmap parameter to 'winter'
)

# Customize the plot
plt.title('Age vs. Cholesterol for Patients Over 50')
plt.xlabel('Age')
plt.ylabel('Cholesterol (chol)')
plt.legend()

# Add mean lines
plt.axvline(mean_age, color='red', linestyle='--', label=f'Mean Age: {mean_age:.2f}')
plt.axhline(mean_chol, color='red', linestyle='--', label=f'Mean Chol: {mean_chol:.2f}')  # Change the color to 'red'

# Save the plot to a file before showing
plt.savefig('heart_disease_plot.png')

# Show the plot
plt.legend()
plt.show()
```
